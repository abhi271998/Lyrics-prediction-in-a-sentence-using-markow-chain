```python
f=open('/content/MyDrive/MyDrive/Transfer_Learning/lyrics2.txt')
ds=f.read()
f.close()

data +=' '.join([ix.strip() for ix in ds.split('\n')]).lower()
print(len(data))
print(data)
```

```python
T={}
memory=4

for ix in range(len(data)-memory):
  ctx=data[ix:ix+memory]
  word=data[ix+memory]

  if T.get(ctx) is None:
    T[ctx]={}
    T[ctx][word]=1
  else:
    if T[ctx].get(word) is None:
      T[ctx][word]=1
    else:
      T[ctx][word]+=1
```

```python
print(T)
```

```python
for kx in T.keys():
  s=float(sum(T[kx].values()))
  for kkx in T[kx].keys():
    T[kx][kkx]=T[kx][kkx]/s
```

```python
print(T)
```

```python
import random
def generate_next(ctx):
  possible_keys=T.get(ctx)
  # print(possible_keys)

  if possible_keys is None:
    return " "

  shuffled_keys=list(possible_keys.keys())
  # print(shuffled_keys)

  random.shuffle(shuffled_keys)
  # print(shuffled_keys)
  
  key_probs=[possible_keys[ix] for ix in shuffled_keys]
  return np.random.choice(shuffled_keys,p=key_probs)
```

```python
print(generate_next('look'))
```

```python
ctx='look'
sentence=''+ctx

for ix in range(1000):
  next_word=generate_next(ctx)
  sentence+=next_word
  ctx=sentence[-memory:]
print(sentence)
```
